```python
# Import the main libraries used for data handling and visualisation.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Loading the Dublin Airbnb Listing dataset from GitHub repository.
url = 'https://github.com/Brithway/Airbnb-Listings-Data-Analyasis/raw/refs/heads/main/Dublin_AirbnbListings.csv'

spf = pd.read_csv(url)
```

```python
# Automatically detect numerical and categorical features.
numerical_cols = spf.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = spf.select_dtypes(include=['object', 'bool']).columns

numerical_cols, categorical_cols
```

```python
# Display the names of identified numerical and categorical columns.
print("Numerical Columns:")
for col in numerical_cols:
    print("-", col)

print("\nCategorical Columns:")
for col in categorical_cols:
    print("-", col)
```

```python
spf[numerical_cols].describe()
```

```python
# Check how many missing values exist in numerical and categorical columns.
print("Missing values in numerical columns:")
print(spf[numerical_cols].isnull().sum())

print("\nMissing values in categorical columns:")
print(spf[categorical_cols].isnull().sum())
```

```python
# Show the first few unique values from each categorical feature.
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(spf[col].unique()[:10])
```

```python
# Convert the 'price' column from text (with € symbol) to numeric values.
spf['price_numeric'] = spf['price'].str.replace('€', '').str.replace(',', '').astype(float)
```

```python
# Check the minimum and maximum prices and count how many listings have extremely high values.
print("Price analysis:")
print(f"Max price: €{spf['price_numeric'].max()}")
print(f"Min price: €{spf['price_numeric'].min()}")

high_prices = spf[spf['price_numeric'] > 1000]
print(f"Listings with price > €1000: {len(high_prices)}")
```

```python
# Plot the distribution of prices between €20 and €500.
reasonable_prices = spf[(spf['price_numeric'] >= 20) & (spf['price_numeric'] <= 500)]

plt.figure(figsize=(14, 8))
sns.histplot(data=reasonable_prices, x="price_numeric",
    kde=True,
    color='violet',
    bins=40)  # More bins for better detail

plt.xlabel('Price (€)', fontsize=16)
plt.ylabel('Number of Listings', fontsize=16)
plt.title('Airbnb Price Distribution in Dublin (€20 - €500)', fontsize=20)
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.show()

print(f"Showing {len(reasonable_prices)} out of {len(spf)} listings (reasonable price range)")
```

```python
# Visualise the distribution of key categorical variables such as room type and neighbourhood.
nominal = ['room_type', 'neighbourhood']
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

for i, col in enumerate(nominal):
    sns.countplot(data=spf,
    x=col,
    color='orange',
    ax=ax[i])

    ax[i].set_xlabel(col, size=12)
    ax[i].set_ylabel('Number of Listings', size=12)
    ax[i].set_title(col + ' Distribution', size=15)
    ax[i].tick_params(axis='x', rotation=45)
```

```python
# Generate boxplots to identify outliers in price and availability Distributions.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Price box plot (filtered)
price_cutoff = spf['price_numeric'].quantile(0.95)
filtered_price = spf[spf['price_numeric'] <= price_cutoff]
sns.boxplot(data=filtered_price, y='price_numeric', ax=ax1, color='lightblue')
ax1.set_ylabel('Price (€)', fontsize=12)
ax1.set_title('Price Distribution\n(Up to 95th percentile)', fontsize=14)

# Availability box plot
sns.boxplot(data=spf, y='availability_365', ax=ax2, color='lightcoral')
ax2.set_ylabel('Availability (Days)', fontsize=12)
ax2.set_title('Availability Distribution', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# Boxplot for minimum nights
min_nights_cutoff = spf['minimum_nights'].quantile(0.90)
filtered_min_nights = spf[spf['minimum_nights'] <= min_nights_cutoff]

plt.figure(figsize=(8, 6))
sns.boxplot(data=filtered_min_nights, y='minimum_nights', color='lightgreen')
plt.ylabel('Minimum Nights', fontsize=14)
plt.title(f'Minimum Nights Distribution\n(Up to {min_nights_cutoff} nights)', fontsize=16)

plt.show()

```

```python
# Boxplot for reviews distribution
reviews_cutoff = spf['number_of_reviews'].quantile(0.90)  # Remove top 10%
filtered_reviews = spf[spf['number_of_reviews'] <= reviews_cutoff]

plt.figure(figsize=(8, 6))
sns.boxplot(data=filtered_reviews, y='number_of_reviews', color='gold')
plt.ylabel('Number of Reviews', fontsize=14)
plt.title(f'Reviews Distribution\n(Up to {reviews_cutoff:.0f} reviews)', fontsize=16)
plt.show()
```

```python
# Calculate maximum, minimum, mode, mean, median, and standard deviation of prices by room type.
room_type_max = spf.groupby('room_type')['price_numeric'].max()
room_type_min = spf.groupby('room_type')['price_numeric'].min()

private_mode = spf[spf['room_type'] == 'Private room']['price_numeric'].mode()
entire_mode = spf[spf['room_type'] == 'Entire home']['price_numeric'].mode()
shared_mode = spf[spf['room_type'] == 'Shared room']['price_numeric'].mode()
hotel_mode = spf[spf['room_type'] == 'Hotel room']['price_numeric'].mode()

room_type_mean = spf.groupby('room_type')['price_numeric'].mean()
room_type_median = spf.groupby('room_type')['price_numeric'].median()
room_type_std = spf.groupby('room_type')['price_numeric'].std()

print('Max:\n', room_type_max)
print('Min:\n', room_type_min)
print('Private Room Mode:\n', private_mode)
print('Entire Home Mode:\n', entire_mode)
print('Shared Room Mode:\n', shared_mode)
print('Hotel Room Mode:\n', hotel_mode)
print('Mean:\n', room_type_mean)
print('Median:\n', room_type_median)
print('Standard Deviation:\n', room_type_std)
```

```python
# Calculate the Interquartile Range (IQR) to identify potential price outliers.
print("IQR Analysis for Dublin City:")
dublin_data = spf[spf['neighbourhood'] == 'Dublin City']['price_numeric']
Q1_dublin = dublin_data.quantile(0.25)
Q3_dublin = dublin_data.quantile(0.75)
IQR_dublin = Q3_dublin - Q1_dublin
max_value_dublin = Q3_dublin + 1.5 * IQR_dublin
min_value_dublin = Q1_dublin - 1.5 * IQR_dublin

print(Q1_dublin)
print(Q3_dublin)
print(IQR_dublin)
print(max_value_dublin)
print(min_value_dublin)

print("\nIQR Analysis for Dn Laoghaire-Rathdown:")
rathdown_data = spf[spf['neighbourhood'] == 'Dn Laoghaire-Rathdown']['price_numeric']
Q1_rathdown = rathdown_data.quantile(0.25)
Q3_rathdown = rathdown_data.quantile(0.75)
IQR_rathdown = Q3_rathdown - Q1_rathdown
max_value_rathdown = Q3_rathdown + 1.5 * IQR_rathdown
min_value_rathdown = Q1_rathdown - 1.5 * IQR_rathdown

print(Q1_rathdown)
print(Q3_rathdown)
print(IQR_rathdown)
print(max_value_rathdown)
print(min_value_rathdown)
```

```python
# To get the High price listing
high_price_listings = spf[spf['price_numeric'] > 200]
print(high_price_listings[['neighbourhood', 'room_type', 'price_numeric', 'minimum_nights', 'number_of_reviews']].head(10))
```

```python
# Scatterplot matrix of key numerical features.
filtered_data = spf[(spf['price_numeric'] <= 500) & (spf['number_of_reviews'] <= 200)]

continuous_features = ['minimum_nights', 'availability_365', 'number_of_reviews', 'reviews_per_month']
fig, ax = plt.subplots(2, 2, figsize=(20, 12))

for i, col in enumerate(continuous_features):
    row = i // 2
    col_idx = i % 2
    sns.scatterplot(data=filtered_data,
    x=col,
    y='price_numeric',
    color='purple',
    alpha=0.7,
    s=80,
    ax=ax[row, col_idx])

    ax[row, col_idx].set_xlabel(col.replace('_', ' ').title(), fontsize=14)
    ax[row, col_idx].set_ylabel('Price (€)', fontsize=14)
    ax[row, col_idx].set_title(f'Price vs {col.replace("_", " ").title()}\n(Filtered: Price ≤ €500)', fontsize=16)
    ax[row, col_idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Analysis of high review Listings
high_review_listings = spf[spf['number_of_reviews'] > 100]
print(f"Listings with more than 100 reviews: {len(high_review_listings)}")
print(high_review_listings[['name', 'neighbourhood', 'room_type', 'price_numeric', 'number_of_reviews']].head(10))
```

```python
# Analysis of highly available listings
high_availability = spf[spf['availability_365'] > 300]
print(f"Listings available more than 300 days per year: {len(high_availability)}")
print(high_availability[['neighbourhood', 'room_type', 'price_numeric', 'availability_365']].head(10))
```

```python
# Correlation analysis between numerical features
numerical_corr = spf[['price_numeric', 'minimum_nights', 'availability_365', 'number_of_reviews', 'reviews_per_month']].corr()
print("Correlation Matrix:")
print(numerical_corr)

# Heatmap of correlations
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Between Numerical Features', fontsize=16)
plt.show()
```

```python
# Price analysis by room type and neighbourhood combination
price_by_room_neighbourhood = spf.groupby(['room_type', 'neighbourhood'])['price_numeric'].mean()
print("Average Price by Room Type and Neighbourhood:")
print(price_by_room_neighbourhood)
```

```python
# Final summary statistics
print("=== FINAL SUMMARY ===")
print(f"Total listings analyzed: {len(spf)}")
print(f"Average price: €{spf['price_numeric'].mean():.2f}")
print(f"Most common room type: {spf['room_type'].mode().iloc[0]}")
print(f"Most common neighbourhood: {spf['neighbourhood'].mode().iloc[0]}")
print(f"Average availability: {spf['availability_365'].mean():.1f} days per year")
print(f"Listings with reviews: {len(spf[spf['number_of_reviews'] > 0])}")
```
